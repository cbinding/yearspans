# Using the Year Span Matcher
## Introduction
The Year Span Matcher determines start/end years from textual expressions of year periods.

```python
from yearspanmatcher.yearspanmatcher import getMatcherForLanguage
from yearspanmatcher.yearspan import YearSpan
from IPython.display import display, HTML
import pandas as pd  # for DataFrame output

# test items
items1 = [
    "January 1066 AD",
    "Early 11th Century AD",
    "Early Eleventh Century BC",
    "early 11th to late 12th century AD",
    "1600-25+17",
    "1600±17",
    "1250 - 57 AD",
    "Edwardian",
    "1850's"
]

items97507 = [
    "romertid",
    "folkevandringstid",
    "førromersk jernalder",
    "yngre bronsealder",
    "nyere tid",
    "folkevandringstid",
    "romertid",
    "folkevandringstid"
]

items71115 = [
    "seinmesolitikum",
    "tidligneolitikum",
    "tidligneolitikum",
    "ca. 4000–3700 f.Kr.",
    "eldre bronsealder",
    "ca. 1400–1050 f.Kr.",
    "eldre jernalder",
    "ca. 250–500 e.Kr.",
    "eldre bronsealder"
]

items71189 = [
    "førromersk jernalder",
    "førromersk jernalder",
    "førromersk jernalder",
    "folkevandringstid"
]



matcher = getMatcherForLanguage("en")

# get year span for each test item
results = []
for item in items1:
    span = matcher.match(item)
    result = {"value": item}
    if (span is not None):
        result["minYear"] = YearSpan.yearToISO8601(span.minYear)
        result["maxYear"] = YearSpan.yearToISO8601(span.maxYear)
        result["isoSpan"] = span.toISO8601()
        results.append(result)

# convert results to pandas dataframe for display  
df = pd.DataFrame(results)
display(df)

```
